# Customer & Inventory Insights
## Day 4 - Behavior Analysis

Load Data

```python
import pandas as pd
```

```python
df = pd.read_csv("cleaned_retail_data.csv")
```

```python
df.head()
```

# CUSTOMER BEHAVIOR

## Average Basket Size

```python
avg_basket = df['quantity'].mean()
print("Average Basket Size:", avg_basket)
```

## Repeat Purchase Pattern

#(Using customer type or invoice)

```python
repeat_customers = df['customer_type'].value_counts()
repeat_customers
```

## Frequently Bought Products

```python
frequent_products = df['product_line'].value_counts()
frequent_products.head()
```

## Visualization

```python
import matplotlib.pyplot as plt
```

```python
frequent_products.head().plot(kind='bar')
plt.title("Top Frequently Bought Products")
plt.xlabel("Product")
plt.ylabel("Count")
plt.show()
```

## INVENTORY ANALYSIS

```python
## Fast-Moving Products
```

```python
fast_moving = df.groupby('product_line')['quantity'].sum().sort_values(ascending=False)
fast_moving.head()
```

```python
## Slow-Moving Products
```

```python
slow_moving = df.groupby('product_line')['quantity'].sum().sort_values()
slow_moving.head()
```

```python
## Dead Stock (Very Low Sales)
```

```python
dead_stock = slow_moving[slow_moving < 50]
dead_stock
```

## Visualization

```python
fast_moving.head().plot(kind='bar')
plt.title("Fast Moving Products")
plt.xlabel("Product")
plt.ylabel("Quantity Sold")
plt.show()
```

## Conclusion

- Identified customer purchasing behavior
- Found frequently bought products
- Detected fast and slow-moving inventory
- Identified potential dead stock
- Insights help improve stock and sales strategy
