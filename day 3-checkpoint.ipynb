# Retail Sales Analysis
## Day 3 - Sales Insights

Load Cleaned Data

```python
import pandas as pd
df = pd.read_csv("cleaned_retail_data.csv")
df.head()
```

## Sales Analysis

```python
total_sales = df['total'].sum()
print("Total Sales:", total_sales)
```

## Average Order Value

```python
avg_order = df['total'].mean()
print("Average Order Value:", avg_order)
```

## Daily Sales Trend

```python
df['date'] = pd.to_datetime(df['date'])
daily_sales = df.groupby('date')['total'].sum()
daily_sales.head()
```

## PLOT

```python
import matplotlib.pyplot as plt
```

```python
daily_sales.plot()
plt.title("Daily Sales Trend")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.show()
```

## PRODUCT ANALYSIS

```python
top_products = df.groupby('product_line')['total'].sum().sort_values(ascending=False)
top_products.head()
```

## Least Selling Products

```python
least_products = df.groupby('product_line')['total'].sum().sort_values()
least_products.head()
```

## CATEGORY ANALYSIS

```python
category_sales = df.groupby('product_line')['total'].sum()
category_sales
```

## Visualization by category

```python
category_sales.plot(kind='bar')
plt.title("Sales by Product Category")
plt.xlabel("Category")
plt.ylabel("Sales")
plt.show()
```

## Conclusion

- Total sales and average order value calculated
- Identified top and least selling product categories
- Observed sales trends over time
- Product categories contribute differently to revenue
