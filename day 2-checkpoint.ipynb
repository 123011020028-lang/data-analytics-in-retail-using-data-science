#Import & Load Data

```python
import pandas as pd
```

```python
df = pd.read_csv("C:\\Users\\DELL\\Downloads\\supermarketsales.csv")
```

```python
df.head()
```

## check missing value

```python
df.isnull().sum()
```

## Handle Missing Values

```python
df = df.dropna() # removes rows with missing values
```

## Remove Duplicates

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

## Fix Date Format

```python
df['Date'] = pd.to_datetime(df['Date'])
```

## Standardize Column Names

```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
df.columns
```

## Clean Text Columns

```python
df['product_line'] = df['product_line'].str.lower()
df['payment'] = df['payment'].str.lower()
```

## Ensure Correct Data Types

```python
df.info()
```

```python
## Save clean dataset
```

```python
df.to_csv("cleaned_retail_data.csv", index=False)
```

## Conclusion

- Missing values removed
- Duplicate records eliminated
- Date column formatted properly
- Column names standardized
- Dataset is clean and ready for analysis
